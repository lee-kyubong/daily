# Scikit learn의 breast cancer dataset을 활용한 '예측모델 성능평가'

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import mean_squared_error
```

```python
data = datasets.load_breast_cancer()
```

```python
type(data)
```

```python
x = data.data
y = data.target
```

```python
import numpy as np
import pandas as pd
```

```python
type(x)
```

```python
x.shape, y.shape
```

```python
y
```

## Holdout

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train, y_train)
#규봉: 데이터 셋을 .fit이라는 함수에 입력하는 것만으로 학습 왼료
#R에 비해 덜 통계지향적, 직관적으로 느껴짐
#R에 비해 더 객체지향적 모습을 보임
```

```python
clf
```

```python
y_pred = clf.predict(x_test)
```

```python
print('Confusion Matrix')
print(confusion_matrix(y_test, y_pred))
#train_test_split에서 random_state에 따라 결과가 다르다
```

```python
print('Accuracy')
print(accuracy_score(y_test, y_pred, normalize = True))
```

```python
print('Classification Report')
print(classification_report(y_test, y_pred))
```

```python
print('AUC: %.2f' % roc_auc_score(y_test, y_pred))
```

```python
print('MSE')
print(mean_squared_error(y_test, y_pred))
```

## k-fold CV

```python
skf = StratifiedKFold(n_splits = 10)
skf.get_n_splits(x, y)
print(skf)
```

```python
for train_index, test_index in skf.split(x, y):
    print('train set: ', train_index)
    print('test set: ', test_index)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf
```

```python
scores = cross_val_score(clf, x, y, cv = skf)
```

```python
print('K-fold CV score')
print(scores)
```

```python
print('Average Accuracy')
print(scores.mean())
```
